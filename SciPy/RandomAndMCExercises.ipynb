## Question 1

The Weibull distribution, with parameters α>0 and β>0 is described by the density function

f(x)={αβ^−α x^(α−1) e^(−(x/β)^α) x>0 , 0  otherwise

(a) Solve for the CDF of the Weibull distribution analytically, using

F(x)=∫ -inf - x  f(t)dt

```python
from scipy.integrate import *
```

```python


#Set up the functions and plot
x = np.arange(0.,20.01,.01)




def Weibull(xlist, alpha=1.5, beta=6.):
    lst = []
    for i in xrange(len(xlist)):
        if xlist[i] <= 0:
            lst.append(0) 
        elif xlist[i] > 0:
            lst.append((alpha/(beta)**alpha)*(xlist[i]**(alpha - 1))*(1/np.exp(((xlist[i]/beta)**alpha))))
    return lst
    
def Weibull_CDF(xlist, alpha=1.5, beta=6.):
    lst = []
    for i in xrange(len(xlist)):
        if xlist[i] <= 0:
            lst.append(0) 
        elif xlist[i] > 0:
            lst.append(1-np.exp(-(xlist[i]/beta)**alpha))
    return lst

plt.plot(x, Weibull(x) ,label=r'$Weibull$')
plt.plot(x, Weibull_CDF(x), label=r'$CDF$')
plt.xlim(0,20)
plt.xlabel(r"$x$",fontsize = 20)
plt.ylabel(r'$f(x),F(x)$',fontsize=20)
plt.legend(loc="upper left",fontsize=15)
plt.show()
```
