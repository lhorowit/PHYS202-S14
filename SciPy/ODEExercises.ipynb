(a) See what happens to the period of the pendulum as you change the initial angle of release. Plot θ vs t for 5 different initial angles, θ0 , spanning 0 to π. (i.e. 0, π/4 , π/2 , 3π/4 , and π ). Put them all on the same plot. What happens to the period of the pendulum as the initial angle increases from π/4 to 3π/4? What happens to the trajectory when the initial angle is 0 and π ?

```python
def deriv(theta_current,t):  # return derivatives of the array theta
    g = 9.8 #m/s^2
    ell = 0.5 #m
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/ell)*np.sin(theta)
    return np.array([ dtheta, domega ])
```

```python
g = 9.8 #m/s^2
ell = 0.5 #m
t = np.arange(0.,50.+0.01,0.01)
omega0 = 0.
th0 = 0.*pi/180 # 0 degrees
th1 = 45.*pi/180 # 45 degrees
th2 = 90.*pi/180 # 90 degrees
th3 = 135.*pi/180 # 135 degrees
th4 = 180.*pi/180 # 180 degrees

theta_app = lambda t: th0*np.cos(np.sqrt(g/ell)*t)


theta_init0 = np.array([th0, omega0]) # initial values
theta_init1 = np.array([th1, omega0])
theta_init2 = np.array([th2, omega0])
theta_init3 = np.array([th3, omega0])
theta_init4 = np.array([th4, omega0])
theta0 = odeint(deriv,theta_init0,t)
theta1 = odeint(deriv,theta_init1,t)
theta2 = odeint(deriv,theta_init2,t)
theta3 = odeint(deriv,theta_init3,t)
theta4 = odeint(deriv,theta_init4,t)


plt.plot(t,theta0[:,0],label="0")
plt.plot(t,theta1[:,0],label="pi/4")
plt.plot(t,theta2[:,0],label="pi/2")
plt.plot(t,theta3[:,0],label="3pi/4")
plt.plot(t,theta4[:,0],label="pi")
plt.xlabel(r'$t (seconds)$',fontsize=20)
plt.ylabel(r'$\theta (rad)$',fontsize=20)
plt.ylim(-4.,4.)
plt.grid()
plt.legend()
plt.show()
```

## Question 2

(a) Use Euler's method (not odeint) to solve the following first-order differential equation with given initial condition y(0) = 0.

dy/dx = x - 2y

Find the value of y when x = 1, y(1), using two different step sizes: 0.25 then 0.02.

```python
def Euler(dt):
    tmax = 100 # maximum time
    nsteps = int(tmax/dt)
    t = np.zeros(nsteps)
    v = np.zeros(nsteps)
    y = np.zeros(nsteps)
 
    # set initial conditions, redundant in this example
    t[0] = 0.
    v[0] = 0.

    for i in range(len(t)-1):
        t[i+1] = t[i] + dt
        v[i+1] = t[i] + 2*y[i]
        y[i+1] = y[i] + dt * v[i]
        
    return t,v,y

t1,dydx1,y1 = Euler(.25)
t2, dydx2, y2 = Euler(.02)
ii = (t1 == 1)
jj = (t1 == 1)
print y1[ii]
print y2[jj]
```

```python
def EulerCromer(dt):
    tmax = 100 # maximum time
    nsteps = int(tmax/dt)
    t = np.zeros(nsteps)
    v = np.zeros(nsteps)
    y = np.zeros(nsteps)
 
    # set initial conditions, redundant in this example
    t[0] = 0.
    v[0] = 0.
    for i in range(len(t)-1):
        t[i+1] = t[i] + dt
        v[i+1] = t[i] + 2*y[i]
        y[i+1] = y[i] + dt * v[i+1]
        
    return t,v,y
```

```python
def Midpoint(dt):
    tmax = 100 # maximum time
    nsteps = int(tmax/dt)
    t = np.zeros(nsteps)
    v = np.zeros(nsteps)
    y = np.zeros(nsteps)
 
    # set initial conditions, redundant in this example
    t[0] = 0.
    v[0] = 0.

    for i in range(len(t)-1):
        t[i+1] = t[i] + dt
        v[i+1] = t[i] + 2*y[i]
        y[i+1] = y[i] + .5 * dt * (v[i] + v[i+1])
        
    return t,v,y
```

```python
t1,dydx1,y1 = Euler(.25)
t2,dydx2,y2 = EulerCromer(.25)
t3,dydx3,y3 = Midpoint(.25)
t4, dydx4, y4 = Euler(.02)
t5, dydx5, y5 = EulerCromer(.02)
t6, dydx6, y6 = Midpoint(.02)
yexact1 = .25*np.exp(2*t1)-.5*t1+.25
yexact2 = .25*np.exp(2*t4)-.5*t4+.25

percentError1 = 100*abs( y1 - yexact1 ) / yexact1
percentError2 = 100*abs( y2 - yexact1 ) / yexact1
percentError3 = 100*abs( y3 - yexact1 ) / yexact1
percentError4 = 100*abs( y4 - yexact2 ) / yexact2
percentError5 = 100*abs( y5 - yexact2 ) / yexact2
percentError6 = 100*abs( y6 - yexact2 ) / yexact2

plt.semilogy(t1, percentError1, label='Euler (.25)')
plt.semilogy(t1, percentError2, label='EulerCromer (.25)')
plt.semilogy(t1, percentError3, label='Midpoint (.25)')
plt.semilogy(t4, percentError4, label='Euler (.02)')
plt.semilogy(t4, percentError5, label='EulerCromer (.02)')
plt.semilogy(t4, percentError6, label='Midpoint (.02)')

plt.axis([0, 10, 1, 100])
plt.xlabel("x",fontsize=20)
plt.ylabel("Percent Error (%)",fontsize=20)
plt.grid()
plt.legend(loc="lower right", fontsize=12)
plt.show()
```
