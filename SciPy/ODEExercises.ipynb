## Question 2

(a) Use Euler's method (not odeint) to solve the following first-order differential equation with given initial condition y(0) = 0.

dy/dx = x - 2y

Find the value of y when x = 1, y(1), using two different step sizes: 0.25 then 0.02.

```python
def Euler(dt):
    tmax = 100 # maximum time
    nsteps = int(tmax/dt)
    t = np.zeros(nsteps)
    v = np.zeros(nsteps)
    y = np.zeros(nsteps)
 
    # set initial conditions, redundant in this example
    t[0] = 0.
    v[0] = 0.

    for i in range(len(t)-1):
        t[i+1] = t[i] + dt
        v[i+1] = t[i] + 2*y[i]
        y[i+1] = y[i] + dt * v[i]
        
    return t,v,y

t1,dydx1,y1 = Euler(.25)
t2, dydx2, y2 = Euler(.02)
ii = (t1 == 1)
jj = (t1 == 1)
print y1[ii]
print y2[jj]
```

```python
def EulerCromer(dt):
    tmax = 100 # maximum time
    nsteps = int(tmax/dt)
    t = np.zeros(nsteps)
    v = np.zeros(nsteps)
    y = np.zeros(nsteps)
 
    # set initial conditions, redundant in this example
    t[0] = 0.
    v[0] = 0.
    for i in range(len(t)-1):
        t[i+1] = t[i] + dt
        v[i+1] = t[i] + 2*y[i]
        y[i+1] = y[i] + dt * v[i+1]
        
    return t,v,y
```

```python
def Midpoint(dt):
    tmax = 100 # maximum time
    nsteps = int(tmax/dt)
    t = np.zeros(nsteps)
    v = np.zeros(nsteps)
    y = np.zeros(nsteps)
 
    # set initial conditions, redundant in this example
    t[0] = 0.
    v[0] = 0.

    for i in range(len(t)-1):
        t[i+1] = t[i] + dt
        v[i+1] = t[i] + 2*y[i]
        y[i+1] = y[i] + .5 * dt * (v[i] + v[i+1])
        
    return t,v,y
```

```python
t1,dydx1,y1 = Euler(.25)
t2,dydx2,y2 = EulerCromer(.25)
t3,dydx3,y3 = Midpoint(.25)
t4, dydx4, y4 = Euler(.02)
t5, dydx5, y5 = EulerCromer(.02)
t6, dydx6, y6 = Midpoint(.02)
yexact1 = .25*np.exp(2*t1)-.5*t1+.25
yexact2 = .25*np.exp(2*t4)-.5*t4+.25

percentError1 = 100*abs( y1 - yexact1 ) / yexact1
percentError2 = 100*abs( y2 - yexact1 ) / yexact1
percentError3 = 100*abs( y3 - yexact1 ) / yexact1
percentError4 = 100*abs( y4 - yexact2 ) / yexact2
percentError5 = 100*abs( y5 - yexact2 ) / yexact2
percentError6 = 100*abs( y6 - yexact2 ) / yexact2

plt.semilogy(t1, percentError1, label='Euler (.25)')
plt.semilogy(t1, percentError2, label='EulerCromer (.25)')
plt.semilogy(t1, percentError3, label='Midpoint (.25)')
plt.semilogy(t4, percentError4, label='Euler (.02)')
plt.semilogy(t4, percentError5, label='EulerCromer (.02)')
plt.semilogy(t4, percentError6, label='Midpoint (.02)')

plt.axis([0, 10, 1, 100])
plt.xlabel("x",fontsize=20)
plt.ylabel("Percent Error (%)",fontsize=20)
plt.grid()
plt.legend(loc="lower right", fontsize=12)
plt.show()
```

Plot the trajectory of the projectile (y vs. x) to see how far it can travel in the x direction before landing.

```python
def Proj_deriv(y,t):
    ay = -9.81 #g in m/s^2
    vx0 = y[0]
    vy0 = y[1]
    vterm = y[2]
    
    
    
    dvxdt = (ay*vx0/vterm**2) * np.sqrt(vy0**2 + vx0**2)
    dvydt = ay * (1 + (vy0/vterm**2) * np.sqrt(vy0**2 + vx0**2))

    return np.array([ dvxdt, dvydt ])
    

def Projectile(v0 = 30, th0 = 50*np.pi/180, time=200, vterm=44):

    
    t = np.arange(0.,time+0.001,0.001)

    #initial conditions
   
    vx0 = v0 * np.cos(th0)
    vy0 = v0 * np.sin(th0)
    x0 = 0
    y0 = 0
    init = np.array([vx0, vy0, vterm])

    #pass the function deriv to the solver with initial values and integration variable
    result = odeint(Proj_deriv, init ,t)

    vx = result[:,0]
    vy = result[:,1]
    print vx
    print vy
    x = vx * t
    y = vy * t
    plt.plot(x,y,label="Quadratic Drag")
    plt.xlabel("x")
    plt.ylabel("y(x)")
    plt.legend()
    plt.axis([0, 50, 0, 20])
    plt.show()
    
Projectile()    
```
